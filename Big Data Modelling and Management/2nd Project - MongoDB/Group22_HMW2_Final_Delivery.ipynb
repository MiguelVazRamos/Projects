Decide what information should be returned in a typical query and optimise the structure for this use case.
```python
# Having a first look at the structure of the documents of our airbnb collection.

pprint(db.listingsAndReviews_new.find_one())
```

```python
# The first thing we decided to do is to create a copy of our original "listingAndReviews_new" collection
# as we will have to change some things on its structure but at the same time we don't want
# to lose its original structure. From here onwards we will be only making changes on this copy called
# "listingAndReviews_work".

query_1 = {
    "$out": "listingsAndReviews_work"
}

pipeline = [query_1]

db.listingsAndReviews_new.aggregate(pipeline)
```

```python
# The first action will be to eliminate the following fields "reviews_copy1", "reviews_copy2",
# "reviews_copy3", and "reviews_copy4". These are copies of the "reviews" field and are considered
# duplicate information that are making the database slower.

query_1 = {
    "$project": {"reviews_copy1": 0,
                "reviews_copy2": 0,
                "reviews_copy3": 0,
                "reviews_copy4": 0}
}

query_2 = {
    "$out": "listingsAndReviews_work"
}

pipeline = [query_1, query_2]

db.listingsAndReviews_work.aggregate(pipeline)
```

```python
# Currently, everytime we show a listing we are presenting too many reviews, and we are also
# presenting information that is not important for the customer, like the "reviewer_id", 
# "reviewer_name". To deal with this, we are first going to apply the "Subset Pattern" so that
# we just keep the first 5 reviews. Then, we are going to apply the "Extended Reference Pattern"
# so that we just present the most important information regarding these 5 reviews to the customer.
```

**- Transformation Applied:** We create a new collection called "reviews" to store the data separately from the main collection.

**- Reason Behind:** We want to have better management and organization for the reviews related data and with this avoid losing important data.

**- Expected Result:** The "reviews" collection should contain individual documents for each review entry with selected fields such as date, comments, listing_id, reviewer_id, and reviewer_name, extracted from the original "reviews" array in the "listingsAndReviews_work" collection.

```python
# Creating a new collection called "reviews" to keep all the data regarding the reviews, as
# this is valuable data and we don't want to lose it.

query_1 = {
    "$unwind":"$reviews"
}

query_2 = {
    "$project":{
        "_id": 0,
        "comments": "$reviews.comments",
        "date": "$reviews.date",
        "listing_id": "$reviews.listing_id",
        "reviewer_id": "$reviews.reviewer_id",
        "reviewer_name": "$reviews.reviewer_name"
    }
}

query_3 = {
    "$out": "reviews"
}

pipeline = [query_1, query_2, query_3]

db.listingsAndReviews_work.aggregate(pipeline)
```

```python
# An example of the type of documents that this collection stores.

pprint(db.reviews.find_one())
```

**- Transformation Applied:** Using the "Extended Reference Pattern" in the "reviews" field from the "listingsAndReviews_work" collection.

**- Reason Behind:** We just want to keep what is important from the field "reviews" to the customer, which is the comments.

**- Expected Result:** We are going to drop from the collection "listingsAndReviews_work" some reviewer information. Nonetheless, it is all stored in the collection "reviews.

```python
# Now, since we are using the "Extended Reference Pattern" we eliminate from the collection
# "listingsAndReviews_work" the field that we don't need. We are not completely losing them
# as they are stored in the collection "reviews".

query_1 = {
    "$project":{
        "reviews.date":0,
        "reviews.listing_id":0,
        "reviews.reviewer_id":0,
        "reviews.reviewer_name": 0
    }
}

query_2 = {
    "$out": "listingsAndReviews_work"}

pipeline = [query_1, query_2]

db.listingsAndReviews_work.aggregate(pipeline)
```

**- Transformation Applied:** Using the "Subset Pattern" in the "reviews" field.

**- Reason Behind:** We just want to keep the first 5 comments from each listing, so that our collection "listingsAndReviews_work" doesn't get very heavy.

**- Expected Result:** We are going to drop from the collection "listingsAndReviews_work" all the comment from the 5th ahead. Nonetheless, we will be able to access them through the collection "reviews".

```python
# As we are using the "Subset Pattern" we are going to only keep the first 5 comments of
# every document. We can do this by using "$set" and "$slice".

query_1 = {
    "$set": {
        "reviews": {"$slice": ["$reviews", 5]}
        }
}

query_2 = {
    "$out": "listingsAndReviews_work"
}

pipeline = [query_1, query_2]

db.listingsAndReviews_work.aggregate(pipeline)
```

```python
# This is how our "reviews" field looks now in the collection "listingsAndReviews_work".

pprint(db.listingsAndReviews_work.find_one()["reviews"])
```

**- Transformation Applied:** Removing "transactions" field from "listingsAndReviews_work" collection.

**- Reason Behind:** The "transactions" field is not needed in the "listingsAndReviews_work" collection as it has been stored separately in the "transactions" collection.

**- Expected Result:** The "transactions" field is removed from the "listingsAndReviews_work" collection, and the data related to transactions is now stored in the "transactions" collection, resulting in a more organized and streamlined structure for the data.

```python
# Now, we still need to deal with the transactions as this is something that we
# don't seE as necessary to show to the possible customers. We are going to store
# their information in a collection called "transactions".

query_1 = {
    "$project": {
        "_id": 0,
        "listing_id": "$_id",
        "transactions": 1
    }
}
    
query_2 = {
    "$out": "transactions"
}

pipeline = [query_1, query_2]

db.listingsAndReviews_work.aggregate(pipeline)
```

```python
# This is what a document from the collection "transactions" looks like.

pprint(db.transactions.find_one())
```

```python
# And now we remove this data from our main collection "listingsAndReviews_work".

query_1 = {
    "$project": {
        "transactions": 0
    }
}
    
query_2 = {
    "$out": "listingsAndReviews_work"
}

pipeline = [query_1, query_2]

db.listingsAndReviews_work.aggregate(pipeline)
```

**- Transformation Applied:** Host information was removed from the original collection "listingsAndReviews_work" and stored in a new collection called "hosts".

```python
# Define the $match stage to filter out documents where price and accommodates fields are missing.
query_1 = {
    "$match": {
        "price": {"$exists": True},
        "accommodates": {"$exists": True}}
    }

# Define the $project stage to create a relevant_difference_provisional field and include the
# accommodates field in the projection. This newly created field calculates the difference
# between the price and the security deposit. In case the security deposit doesn't exist,
# it will assume it to be 0.
query_2 = {   
    "$project": {
        "_id": 1,
        "relevant_difference_provisional": {
            "$subtract": [
                "$price",
                {"$ifNull": ["$security_deposit", 0]}]},
        "accommodates": 1}
    }

# Define the second $project stage to create a relevant_difference_final field by dividing
# the relevant_difference_provisional field by accommodates and converting the result to a double.
# This is important as the question asks us to take into consideration the number of accomodates.
query_3= {
    "$project": {
        "_id": 1,
        "relevant_difference_final": {
            "$toDouble": {
                "$divide": [
                    "$relevant_difference_provisional", "$accommodates"]}}}
    }

# Define the $sort stage to sort the documents by the relevant_difference_final field
# in descending order.
query_4 = {"$sort": {"relevant_difference_final": -1}}

pipeline = [query_1, query_2, query_3, query_4]

iterator = db.listingsAndReviews_work.aggregate(pipeline)

result=list(iterator)

# Defining the maximum value.
max_instance=result[0]
max_value=float(max_instance["relevant_difference_final"])
pprint(f"The biggest difference is {max_value}.")

# Defining the minimum value.
min_instance=result[-1]
min_value=float(min_instance["relevant_difference_final"])
pprint(f"The smallest difference is {min_value}.")
```

```python
# Creating a function that will count the reviews according to the "reviewer_id" or
# the "reviewer_name". As the "reviewer_id" is unique, it will return only the numbers
# of reviews written of that specific person. But since we just have the first name
# of each one of the reviewers the function will return how many reviews each person
# with the specified name written.

def review_counter(reviewer_id=None, reviewer_name=None):
    
# Creating a if condition when the user doesn't insert anything.    
    if reviewer_id is None and reviewer_name is None:
        raise ValueError("At least one parameter must be provided")
        
# Defining the first step of our query when the "reviewer_id" is specified.
    elif reviewer_id is None and reviewer_name is not None:
        query_1={"$match": {"reviewer_name": reviewer_name}}

# Defining the first step of our query when the "reviewer_name" is specified.
    elif reviewer_id is not None and reviewer_name is None:
        query_1={"$match": {"reviewer_id": reviewer_id}}

# Defining the rest of the query.
    query_2={
        "$group": {
            "_id": {"reviewer_id": "$reviewer_id",
                    "reviewer_name": "$reviewer_name"
                   },
            "count": {"$sum": 1}}
        }    
    
    query_3={
        "$project": {
            "_id": 0,
            "reviewer_id": "$_id.reviewer_id",
            "reviewer_name": "$_id.reviewer_name",
            "count": "$count"}
        }    
    
    pipeline=[query_1, query_2, query_3]
    
    iterator = db.reviews.aggregate(pipeline)

    document = list(iterator)

    return document
```

```python
# Taking out the field "review_scores" out of our main collection "listingsAndReviews_work".

query_1 = {
    "$project": {
        "review_scores": 0
    }
}

query_2 = {
    "$out": "listingsAndReviews_work"
    }

pipeline = [query_1, query_2]

db.listingsAndReviews_work.aggregate(pipeline)
```

```python
# We are going to start our query by using "$objectToArray", this aggregation expression
# is useful as it will transform each score and its name into an array of key-value pairs.
# It is useful when you need to iterate over the keys and values of a document, as it
# transforms the document into an array that is easier to work with in the aggregation pipeline.
query_1 = {
        "$addFields": {
            "all_review_scores": {"$objectToArray": "$review_scores"}
    }
}

# Taking out the previous "review_scores" field, as now we have the "all_review_scores"
# that has the array of key-value pairs.
query_2 = {
    "$project": {
        "review_scores":0
    }
    
}

# Unwiding each one of the scores, as each array has several elements.
query_3 = {
    "$unwind": "$all_review_scores"
}

# Then we are grouping by "listing_id" and summing all the values of our key-value pairs
# so that we get a new metric called "total_scores", we will then use it later to calculate
# the average.
query_4 = {
    "$group": {
        "_id":"$listing_id",
        "total_scores": {"$sum": "$all_review_scores.v"},
        "count": {"$sum": 1}
    }
}

# Cleaning slightly the ouput to be more presentable.
query_5 = {
    "$project": {
        "listing_id": "$_id",
        "_id": 0,
        "average_review_scores": {"$divide": ["$total_scores", "$count"]}
    }
}

pipeline = [query_1, query_2, query_3, query_4, query_5]

iterator = db.review_scores.aggregate(pipeline)

result = list(iterator)

pprint(result)
```

```python
# Additionally, and not a part of the exercise we are going to create a function
# that given a "listing_id" it will output the "average_review_scores". Later during
# Question 15 it will allows use to see that everytime a metric of score is added to
# a document of our "score_reviews" collection it will be included on the calculation
# of the "average_review_scores" we ouputed above.
```

```python
# This function will be very useful during Question 15.

def calculate_average_metric(listing_id):
    
# The only difference from the previous presented query is that here we will select
# the "listing_id" we want to look into to.
    query_0 = {
        "$match": {"listing_id": listing_id}
    }

    query_1 = {
            "$addFields": {
                "all_review_scores": {"$objectToArray": "$review_scores"}
        }
    }

    query_2 = {
        "$project": {
            "review_scores":0
        }

    }

    query_3 = {
        "$unwind": "$all_review_scores"
    }


    query_4 = {
        "$group": {
            "_id":"$listing_id",
            "total_scores": {"$sum": "$all_review_scores.v"},
            "count": {"$sum": 1}
        }
    }

    query_5 = {
        "$project": {
            "listing_id": "$_id",
            "_id": 0,
            "average_review_scores": {"$divide": ["$total_scores", "$count"]}
        }
    }

    pipeline = [query_0, query_1, query_2, query_3, query_4, query_5]

    iterator = db.review_scores.aggregate(pipeline)

    result = list(iterator)

    return result
```

11) We wish to be able to have better access to information about transaction, we wish to develop a search engine that can calculate the average value of transactions in a given period of time quickly for a given property.

```python
# Creating a composite index as on the function we are going to create we are going
# to use the operator "$match" to match specific "listing_id" and "transactions.transactions.date".

# Although we will start our function with the "$unwind" operator, these indexes will still have an
# effect. The MongoDB query planner can still be effective in optimizing the query in query_2.

db.collection.create_index([("listing_id", 1), ("transactions.transactions.date", 1)])
```

```python
# The following function will calculate the average value of the transactions
# when given a period of time.

def average_transaction_time(listing_id, lower_limit, higher_limit):
    
# We start by unwiding the transactions as we have several transactios for the
# same listing.
    query_1 = {
        "$unwind": "$transactions.transactions"
    }

# Then, we match for the specific "listing_id" we are looking for and also we only
# include transactions that occured between the period of time we specified in
# the input of our function.
    query_2 = {
        "$match": {
                "listing_id": listing_id,
                "transactions.transactions.date": {"$gte": lower_limit,
                                                  "$lte": higher_limit}
        }
    }

# Then we calculate the sum of all the transaction values for that specific "listing_id",
# and also the total number of transactions so that later we can calculate the average
# as requested in the question.
    query_3 = {
        "$group": {
            "_id": "$listing_id",
            "sum": {"$sum": {"$toDouble": "$transactions.transactions.price"}},
            "count": {"$sum": 1}
        }
    }

# Finally calculating the average value taking into consideration the total value and number
# of transactions within the defined period.
    query_4 = {
        "$project": {
            "_id": 0,
            "listing_id": "$_id",
            "avg_price": {"$divide": ["$sum", "$count"]}
        }
    }


    pipeline = [query_1, query_2, query_3, query_4]

    iterator = db.transactions.aggregate(pipeline)

    result = list(iterator)

    return result
```

```python
# Here is an example on how to use the created function.

listing_id_example = "10006546"
lower_limit_example = datetime.datetime(2008, 8, 12, 0, 0)
higher_limit_example = datetime.datetime(2015, 12, 17, 0, 0)

average_transaction_time(listing_id_example, lower_limit_example, higher_limit_example)
```

12) We wish to have a summary webpage that displays information about our top destinations. This webpage should display for each of the top 10 cities some basic information about our operations in the area (number of properties by type for example, average price by type) but you can choose the metrics. For each of the top 10 cities it should also provide some basic information about the top 3 properties in each city (price, number of review, whatever you think useful) to show an example of the properties available in the area. We would like to keep this webpage up to date as information changes.

```python
# We are going to divide this question in two sub-questions. Firstly we will present regarding 
# the top 10 cities, the number of properties and average price by property type. Then, we are going to
# present the top 3 properties (based on > 20 reviews and highest rating), their price and the
# number of reviews.
```

```python
# Creating an index in the field "address.market" from the "listingsAndReviews_work" collection, as
# it will facilitate the group by we are doing in the first step of our pipeline.

db.listingsAndReviews_work.create_index("address.market")
```

```python
# In this cell we will be asnwering the top 10 cities based on the number of listings and
# then present the number of properties and average price by property type for each one
# of these top 10 cities. In the first three queries we are going to find which are the
# top 10 cities based on the number of listings.

# Grouping by each city and counting how many listings each one of them have.
query_1 = {
    "$group": {
        "_id": "$address.market",
        "count": {"$sum": 1}
    }
}

# Sorting in a descending order.
query_2 = {
    "$sort": {"count": -1}
}

# Limiting the output to 10, as we only want the top 10 cities.
query_3 = {
    "$limit": 10
}

# Then for each one of the top 10 cities we are going to join all the listing information
# we have from the "listingsAndReviews_work". We will create a field "listings" where we
# are storing all the information regarding listings for that particular city from the
# top 10 cities.
query_4 = {
    "$lookup": {
        "from": "listingsAndReviews_work",
        "localField": "_id",
        "foreignField": "address.market",
        "as": "listings"
    }
}

# Unwiding the listings, as there are several listings for each one of the top 10 cities.
query_5 = {
    "$unwind": "$listings"
    }

# Grouping by the city name and the type of property, during the group by we will be
# calculating the total price of these listings as well as the number of listings,
# so that later we can calculate the average.
query_6 = {
    "$group":{
        "_id": {
            "_id": "$_id",
            "property_type": "$listings.property_type"
                },
            "total_price": {"$sum": "$listings.price"},
            "frequency": {"$sum": 1 },
        }
    }

# Cleaning our final output by using the "$project" method.
query_7 = {
    "$project": {
        "_id":0,
        "number_of_listings": "$frequency",
        "city": "$_id._id",
        "property_type": "$_id.property_type",
        "average_price": {"$divide": ["$total_price", "$frequency"]}
    }
}

pipeline = [query_1, query_2, query_3, query_4, query_5, query_6, query_7]

iterator = db.listingsAndReviews_work.aggregate(pipeline)

result = list(iterator)

pprint(result)
```

```python
# In this cell, we are going to present the top 3 properties for each one of the top
# 10 cities. We will decide which are the top 3 properties based on the highest rating
# for the cases where the number os reviews is higher then 20. Then we will present the price
# and number of reviews for these properties.

# Grouping by each city and counting how many listings each one of them have.
query_1 = {
    "$group": {
        "_id": "$address.market",
        "count": {"$sum": 1}
    }
}

# Sorting in a descending order.
query_2 = {
    "$sort": {"count": -1}
}

# Limiting the output to 10, as we only want the top 10 cities.
query_3 = {
    "$limit": 10
}

# Then for each one of the top 10 cities we are going to join all the listing information
# we have from the "listingsAndReviews_work". We will create a field "listings" where we
# are storing all the information regarding listings for that particular city from the
# top 10 cities.
query_4 = {
    "$lookup": {
        "from": "listingsAndReviews_work",
        "localField": "_id",
        "foreignField": "address.market",
        "as": "listings"
    }
}

# Unwiding the listings, as there are several listings for each one of the top 10 cities.
query_5 = {
    "$unwind": "$listings"
}

# Until here what we did was exactly what was explained in the previous cell, going ahead
# we will have to change sligthly the code. We will have to merge with the documents of
# the "review_scores" so that we have access to the field "review_scores_rating" as we
# will use it as our ultimate rating.


query_6 = {
    "$lookup": {
        "from": "review_scores",
        "localField": "listings._id",
        "foreignField": "listing_id",
        "as": "review_scores"
    }
}

# We use "$project" to keep only the data we need.
query_7 = {
    "$project": {
        "_id": 0,
        "city": "$_id",
        "listing_id": "$listings._id",
        "price": "$listings.price",
        "number_reviews": "$listings.number_of_reviews",
        "rating": "$review_scores.review_scores.review_scores_rating"
    }
}

# Unwiding rating as its value came stored in a list. Just that,
# because each listing only has one rating assigned to them.
query_8 = {
    "$unwind": "$rating"
}

# Keeping only the listings that have more than 20 reviews, we understood
# that less than this it wouldn't be representative enough.
query_9 = {
    "$match": {
        "number_reviews": {"$gte": 20}
    }
}

# Sorting in a desceding manner as we want the highest ratings.
query_10 = {
    "$sort": {"rating": -1}
}

# Grouping by city and using $push to add all the listings from that city
# to an array called "listings".
query_11 = {
    "$group": {
        "_id": "$city",
        "listings": {"$push": "$$ROOT"}
    }
}

# Slicing to only keep the top three ratings of every city.
query_12 = {
    "$project": {
        "_id":0,
        "city": "$_id",
        "listings": {"$slice": ["$listings", 3]}
    }
}

# Unwiding the three listings we have stored in the array "listings"
# for each one of the top three cities.
query_13 = {
    "$unwind": "$listings"
}

# Projecting our final output in a more presentable way.
query_14 = {
    "$project": {
        "city": "$city",
        "listing_id": "$listings.listing_id",
        "number_reviews": "$listings.number_reviews",
        "price": "$listings.price",
        "rating": "$listings.rating"
    }
}

pipeline = [query_1, query_2, query_3, query_4, query_5, query_6, query_7, query_8, query_9, query_10, query_11, query_12, query_13, query_14]

iterator = db.listingsAndReviews_work.aggregate(pipeline)

result = list(iterator)

pprint(result)
```
